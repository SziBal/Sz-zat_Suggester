```python
def solve(testWord,inp,lista):
    testWordC=list(set(testWord))
    posA = []
    posL = []
    for letter in testWordC:
        pos = []
        for j,letterT in enumerate(testWord):
            if letter==letterT:
                pos.append(j)

        posI = len(pos)
        posA.append(pos)
        posL.append([inp[index] for index in pos])
    

    toremove = []
    for word in lista:
        added = False
        i=0
        while added == False and i<5:
            if inp[i]=="_" or inp[i]=="?":
                if word[i]==testWord[i] and word not in toremove:
                    toremove.append(word)
                    added = True
            else:
                if word[i]!=testWord[i] and word not in toremove:
                    toremove.append(word)
                    added = True
            i=i+1

        i = 0
        while added == False and i<len(posL):
            if posL[i].count("_")>0 and word.count(testWordC[i]) != (posL[i].count("?")+posL[i].count("+")):
                toremove.append(word)
                added = True
            elif posL[i].count("_")==0 and word.count(testWordC[i]) < (posL[i].count("?")+posL[i].count("+")):
                toremove.append(word)
                added = True
            i=i+1
    [lista.remove(rem) for rem in toremove]
```

```python
def suggestionA(lista):
    possible = []
    for item in lista:
        worst=0
        for product in itertools.product("_+",repeat=5):
            inp2 = "".join(list(product))
            listb=lista.copy()

            solve(item,inp2,listb)

            if len(listb) > worst:
                worst=len(listb)
        possible.append([item,worst])

    sortedPoss = sorted(possible, key=lambda x: x[1], reverse=True)
    return [["".join(row[0]),row[1]] for row in sortedPoss]
```
